```python
ocr = PaddleOCR(use_angle_cls=True, lang='en',use_gpu=True)

# Function to download and open image from URL
def load_image_from_url(url):
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))
    #grayscale_image = ImageOps.grayscale(img)
    return np.array(img)

# Helper function to calculate distance in both horizontal and vertical directions (for fixed threshold)
def is_nearby_fixed_threshold(box1, box2, threshold=100000):
    x1, y1, x2, y2 = box1[0][0], box1[0][1], box1[2][0], box1[2][1]
    a1, b1, a2, b2 = box2[0][0], box2[0][1], box2[2][0], box2[2][1]

    # Calculate distances
    horizontal_dist = min(abs(x1 - a2), abs(x2 - a1))  # Left/Right proximity
    vertical_dist = min(abs(y1 - b2), abs(y2 - b1))    # Above/Below proximity

    # Check if either distance is within the threshold
    return horizontal_dist < threshold or vertical_dist < threshold

# Helper function to calculate distance in both directions based on image resolution
def is_nearby_by_image_resolution(box1, box2, image_width, image_height, threshold_ratio=5):
    adaptive_threshold = max(image_width, image_height) * threshold_ratio
    x1, y1, x2, y2 = box1[0][0], box1[0][1], box1[2][0], box1[2][1]
    a1, b1, a2, b2 = box2[0][0], box2[0][1], box2[2][0], box2[2][1]

    # Calculate distances
    horizontal_dist = min(abs(x1 - a2), abs(x2 - a1))  # Left/Right proximity
    vertical_dist = min(abs(y1 - b2), abs(y2 - b1))    # Above/Below proximity

    # Check if either distance is within the threshold
    return horizontal_dist < adaptive_threshold or vertical_dist < adaptive_threshold

# Helper function to calculate distance based on text height in both directions
def is_nearby_by_text_size(box1, box2, scale_factor=5):
    box1_height = box1[2][1] - box1[0][1]
    box2_height = box2[2][1] - box2[0][1]
    average_height = (box1_height + box2_height) / 2
    adaptive_threshold = average_height * scale_factor

    x1, y1, x2, y2 = box1[0][0], box1[0][1], box1[2][0], box1[2][1]
    a1, b1, a2, b2 = box2[0][0], box2[0][1], box2[2][0], box2[2][1]

    # Calculate distances
    horizontal_dist = min(abs(x1 - a2), abs(x2 - a1))  # Left/Right proximity
    vertical_dist = min(abs(y1 - b2), abs(y2 - b1))    # Above/Below proximity

    # Check if either distance is within the threshold
    return horizontal_dist < adaptive_threshold or vertical_dist < adaptive_threshold

def draw_selected_boxes(image, ocr_result, extracted_data):
    draw = ImageDraw.Draw(image)

    # Iterate over OCR result to find and draw only numerical values with nearby text
    for line in ocr_result:
        for word_info in line:
            text = word_info[1][0]
            if text in extracted_data:
                box = word_info[0]
                draw.polygon([tuple(point) for point in box], outline="green", width=3)  # Draw bounding box

    return image

def get_bounding_box_dimensions(box):
    # Calculate distances between points to determine width and height
    # box = [[x1, y1], [x2, y2], [x3, y3], [x4, y4]]

    width = np.linalg.norm(np.array(box[0]) - np.array(box[1]))  # Distance between point 1 and point 2 (width)
    height = np.linalg.norm(np.array(box[1]) - np.array(box[2]))  # Distance between point 2 and point 3 (height)

    # Return the width and height
    return width, height

# Perform OCR and extract relevant information
# Perform OCR and extract relevant information
def extract_numerical_values(image, method="fixed"):
    image_width, image_height = image.shape[1], image.shape[0]  # Get image dimensions for scaling
    result = ocr.ocr(image, cls=True)

    # Process detected text lines
    extracted_data = []
    for line in result:
        for idx, word_info in enumerate(line):
            text = word_info[1][0]
            if any(char.isdigit() for char in text):  # If text contains digits
                box = word_info[0]

                # Get dimensions of the bounding box
                width, height = get_bounding_box_dimensions(box)

                # Add to extracted data only if the bounding box width is greater than the height
                if width > height:
                    combined_text = text  # Only keep the number

                    # Check if there's a nearby word to combine
                    if idx + 1 < len(line):
                        next_word_info = line[idx + 1]
                        next_text = next_word_info[1][0]

                        # Choose method to check proximity in both directions
                        if method == "fixed":
                            nearby = is_nearby_fixed_threshold(word_info[0], next_word_info[0])
                        elif method == "resolution":
                            nearby = is_nearby_by_image_resolution(word_info[0], next_word_info[0], image_width, image_height)
                        elif method == "text_size":
                            nearby = is_nearby_by_text_size(word_info[0], next_word_info[0])
                        else:
                            raise ValueError("Unknown method for proximity check")

                        # Check if the next word is near the current word in any direction (left, right, above, below)
                        if nearby and next_text.isalpha():
                            combined_text = combined_text + " " + next_text

                    # Append the extracted number (and nearby word if found) to the list
                    extracted_data.append(combined_text)

    return extracted_data, result  # Return both extracted data and the OCR result for further use




def process_image(image_url):
    # Load the image from the URL
    image = load_image_from_url(image_url)

    # Perform OCR and extract numerical values with nearby text
    extracted_data, ocr_result = extract_numerical_values(image, method="fixed")

    return extracted_data, ocr_result

# Function to plot the selected bounding boxes on the image
def plot_selected_boxes(image_url, ocr_result, extracted_data):
    # Re-load the image to draw on it (as PIL Image object)
    image_pil = Image.open(BytesIO(requests.get(image_url).content))

    # Draw only the selected boxes (numerical values + nearby text)
    image_with_selected_boxes = draw_selected_boxes(image_pil, ocr_result, extracted_data)

    # Display the image with the selected bounding boxes
    plt.figure(figsize=(10, 10))
    plt.imshow(image_with_selected_boxes)
    plt.axis("off")
    plt.show()
```
